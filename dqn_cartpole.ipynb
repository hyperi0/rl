```python
import gymnasium as gym
from stable_baselines3 import A2C
```

```python
env = gym.make("CartPole-v1", render_mode="rgb_array")
```

```python
model = A2C("MlpPolicy",
            env=env,
            verbose=1,
            tensorboard_log="./logs/dqn_cartpole_tensorboard")
model.learn(total_timesteps=10000,
            callback = )
model.save("./agents/dqn_cartpole")
```

```python
import time

env = model.get_env()

ep_rewards = []
eps = 3

for i in range(eps):

    total_reward = 0
    obs = env.reset()
    done = False

    while not done:
        action, _state = model.predict(obs, deterministic=True)
        obs, reward, done, info = env.step(action)
        total_reward += reward
        env.render("human")

    print("Total reward: " + str(total_reward))
    time.sleep(1)
    
```
